```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# ippo vs mappo
data_path1 = "completed_trainings/IPPOTrainer_mpe_simple_adversary_3458f_00000_0_2024-01-24_15-09-36/progress.csv"
data_path2 = "completed_trainings/MAPPOTrainer_mpe_simple_adversary_81809_00000_0_2024-01-24_15-47-33/progress.csv"
```

```python
df1 = pd.read_csv(data_path1)
df2 = pd.read_csv(data_path2)
```

```python
plt.plot(df1.episode_reward_mean, label="IPPO")
plt.plot(df2.episode_reward_mean, label="MAPPO")
plt.axhline(0, color='black', linestyle='--', linewidth=1, label='y=0')

plt.xlabel('Epoch')
plt.ylabel('Average reward')
plt.title('IPPO vs MAPPO')
plt.legend()
```

```python
plt.plot(df1["info/learner/policy_agent_/learner_stats/vf_explained_var"], label="IPPO")
plt.plot(df2["info/learner/policy_agent_/learner_stats/vf_explained_var"], label="MAPPO")
plt.axhline(0, color='black', linestyle='--', linewidth=1, label='y=0')

plt.xlabel('Epoch')
plt.ylabel('Variance explained')
plt.title('IPPO vs MAPPO')
plt.legend()
```
