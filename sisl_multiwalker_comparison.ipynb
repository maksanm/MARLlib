# IPPO vs MAPPO in MPE Simple Adversary multi-agent environment

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
mappo_data_path = "results\mappo_mlp_multiwalker\progress.csv"
ippo_data_path = "results\ippo_mlp_multiwalker\progress.csv"
```

```python
mappo_df = pd.read_csv(mappo_data_path)
ippo_df = pd.read_csv(ippo_data_path)
```

```python
mappo_df
```

```python
ippo_df
```

```python
plt.plot(mappo_df.episode_reward_mean, label="MAPPO")
plt.plot(ippo_df.episode_reward_mean, label="IPPO")
plt.axhline(0, color='black', linestyle='--', linewidth=1, label='y=0')

plt.xlabel('Epoch')
plt.ylabel('Average reward')
plt.title('IPPO vs MAPPO')
plt.legend()
```

```python
plt.plot(mappo_df["info/learner/shared_policy/learner_stats/total_loss"], label="MAPPO")
plt.plot(ippo_df["info/learner/shared_policy/learner_stats/total_loss"], label="IPPO")
plt.axhline(0, color='black', linestyle='--', linewidth=1, label='y=0')

plt.xlabel('Epoch')
plt.ylabel('Variance explained')
plt.title('IPPO vs MAPPO')
plt.legend()
```
